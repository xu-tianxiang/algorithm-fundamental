# 一维傅里叶变换

```python
import numpy as np
import matplotlib.pyplot as plt

A = 1
t = np.linspace(0, 2*np.pi, 1000)
x = A * np.sin(2*t)

plt.plot(t, x, 'b-')
plt.show()
```

```python
fourier_transform = np.fft.fft(x)
power_specturm =np.abs(fourier_transform)/len(x)
plt.plot(power_specturm)
plt.show()
```

```python
max_sig = np.max(x)
min_sig = np.min(x)
x = x/((max_sig - min_sig)/2)
```
